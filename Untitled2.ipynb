```python
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
import yfinance as yf
import seaborn as sns
```

```python
start = yf.download('TCS.NS',
                   start='2020-01-01',
                   end='2024-01-01',
                   progress='false')
start
```

```python
start.info()
```

```python
start['Close'].plot(figsize=(12,8))
```

```python
start['Volume'].plot(figsize=(12,8))
```

```python
start[['Close','Volume']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start.describe()
```

```python
start.isnull().sum()
```

```python
start.isnull()
```

```python
start['simple_rtn']=start.Close.pct_change()
start
```

```python
start['Volume_rtn']=start.Volume.pct_change()
start
```

```python
start[['simple_rtn','Volume_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start.tail(10)
```

```python
start['log_rtn']=np.log(start.Close/start.Close.shift(1))
start
```

```python
start['log_rtn']=np.log(start.Volume/start.Close.shift(1))
start
```

```python
start[['log_rtn','Volume_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start[['simple_rtn','log_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
Window_Size=12
start['rolling_mean']=start.Close.rolling(window=Window_Size).mean()
start['rolling_std']=start.Close.rolling(window=Window_Size).std()
start
```

```python
Window_Size=12
start['Vrolling_mean']=start.Volume.rolling(window=Window_Size).mean()
start['Vrolling_std']=start.Volume.rolling(window=Window_Size).std()
start
```

```python
start[['Vrolling_mean','rolling_mean']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start['logrolling_mean']=np.log(start.rolling_mean/start.rolling_mean.shift(1))
start['logrolling_std']=np.log(start.rolling_std/start.rolling_std.shift(1))
start
```

```python
start[['logrolling_mean','logrolling_std']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start['logVrolling_mean']=np.log(start.Vrolling_mean/start.Vrolling_mean.shift(1))
start['logVrolling_std']=np.log(start.Vrolling_std/start.Vrolling_std.shift(1))
start
```

```python
start[['logVrolling_mean','logVrolling_std']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
start[['logrolling_mean','logVrolling_mean']].plot(subplots=True,style='b',figsize=(12,8))
```
