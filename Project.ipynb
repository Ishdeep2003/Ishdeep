```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Load your dataframe
# df = pd.read_csv("your_file.csv")

# 1. Create classification target (Up/Down)
df["target"] = (df["close "].shift(-1) > df["close "]).astype(int)

# Drop the last row (NaN target due to shift)
df = df.dropna()

# 2. Features (X) and target (Y)
X = df.drop(["Date ", "series ", "target"], axis=1)  # drop non-numeric
Y = df["target"]

# If you still have object columns, encode them
for col in X.select_dtypes(include=["object"]).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# 3. Apply SMOTE
smote = SMOTE(random_state=42)
X_resample, Y_resample = smote.fit_resample(X, Y)

print("Before SMOTE:", Y.value_counts().to_dict())
print("After SMOTE:", pd.Series(Y_resample).value_counts().to_dict())
```
